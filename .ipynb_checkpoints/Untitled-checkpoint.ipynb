```python
ones=np.ones((X_train.shape[0],1))
X_train=np.hstack((ones,X_train))
print(X_train.shape)
type(X_train)
```

# Implementing Linear Regression

```python
def hypothesis(x, theta):
    y_=0.0
    n=x.shape[0]
    for i in range(n):
        y_+=theta[i]*x[i]
    return y_

def error(X,Y,theta):
    
    m = X.shape[0]
    e = 0.0 
    for i in range(m):
        y_ = hypothesis(X[i], theta)
        e += (y_ - Y[i])**2
        
    return e/m

def gradient(X,Y,theta):
    m,n=X.shape
    grad = np.zeros((n,))
    for j in range(n):
        for i in range(m):
            y_ = hypothesis(X[i],theta)
            grad[j] += (y_ - Y[i])*X[i][j]
            
    return grad/m
    
def gradientDescent(X,Y,learning_rate=0.05,max_epoch=900):
    
    m,n=X.shape
    theta = np.zeros((n,))
    error_list = []
    
    for i in range(max_epoch): 
        e = error(X,Y,theta)
        error_list.append(e)
        grad = gradient(X,Y,theta)
        for j in range(n):
            theta[j] = theta[j] - learning_rate*grad[j]
        
    return theta, error_list
    
```

```python
final_theta, error_list = gradientDescent(X_train,Y_train)
```

```python
plt.plot(error_list)
plt.show()
```

# In the above graph we can see our how error is decreasing over time.

```python
print(final_theta)
```

```python
print(error_list)
```

# Prediction 

```python
y_pred=[]
m=X_train.shape[0]

for i in range(m):
    pred=hypothesis(X_train[i],final_theta)
    y_pred.append(pred)
```

```python
y_pred
```

# Calculating Accuracy

```python
def r2_score(Y_train,y_pred):
    num=np.sum((Y_train-y_pred)**2)
    denom=np.sum((Y_train-Y_train.mean())**2)
    score=(1-num/denom)
    return score*100
```

```python
accuracy = r2_score(Y_train, y_pred)  
print("Training accuracy:", accuracy)
```
